Binary Classification

Given an age of a person, come up with a **function** that can predict if person will buy insurance or not.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = np.array([22, 25, 47, 52, 46, 56, 55, 60, 62, 61, 18, 28, 27, 29, 49])
have_insurance = np.array([0, 0, 1, 0, 1, 1, 0, 1, 1, 1, 0, 0, 0, 0, 1])
```

```python
plt.scatter(age, have_insurance)
plt.show()
```

```python
# y = a*x + b
a, b = np.polyfit(age, have_insurance, 1)
print(f"Coeficiente a: {a} | Coeficiente b: {b}")
```

```python
plt.scatter(age, have_insurance)
plt.plot(age, age * a + b, color="red") 
plt.plot(age, age*0.03 - 0.7) # Aproximação feito a mão
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def prediction_function(age):
    z = a * age + b # 0.04150133 ~ 0.042 and -1.52726963 ~ -1.53
    y = sigmoid(z)
    return y
```

```python
pred = prediction_function(age)

plt.scatter(age, have_insurance)
plt.plot(age, pred, color="red")
plt.plot(age, sigmoid(age), color="green")
plt.plot(age, age*a + b) 
plt.show()
```
